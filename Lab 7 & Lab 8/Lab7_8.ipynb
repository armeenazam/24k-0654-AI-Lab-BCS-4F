```python
#task1
def minimax(node, depth, isMax, tree, path=""):
    if depth == 0:
        print(f"Visited leaf {path} → {node}")
        return node

    if isMax:
        best = float('-inf')
        for i, child in enumerate(tree):
            val = minimax(child, depth-1, False, child if isinstance(child, list) else child, path + f"->{i}")
            best = max(best, val)
        return best
    else:
        best = float('inf')
        for i, child in enumerate(tree):
            val = minimax(child, depth-1, True, child if isinstance(child, list) else child, path + f"->{i}")
            best = min(best, val)
        return best

tree = [
    [[4,7],[2,5]],
    [[1,8],[3,6]]
]
print("Minimax Value:", minimax(tree, 3, True, tree))
```

```python
#task2
def alphabeta(node, depth, alpha, beta, isMax):
    if depth == 0:
        return node

    if isMax:
        value = float('-inf')
        for child in node:
            value = max(value, alphabeta(child, depth-1, alpha, beta, False))
            alpha = max(alpha, value)
            if alpha >= beta:
                print("Pruned at Max node")
                break
        return value
    else:
        value = float('inf')
        for child in node:
            value = min(value, alphabeta(child, depth-1, alpha, beta, True))
            beta = min(beta, value)
            if alpha >= beta:
                print("Pruned at Min node")
                break
        return value

tree = [
    [[4,7],[2,5]],
    [[1,8],[3,6]]
]

print("Alpha-Beta Value:", alphabeta(tree, 3, float('-inf'), float('inf'), True))
```

```python
#task3

tree = [
    [[4, 7], [2, 5]],
    [[1, 8], [3, 0]]
]

def minimax(node, depth, isMax, name="Root"):
    if depth == 0:
        print(f"{name} = {node}")
        return node, []

    if isMax:
        best = float('-inf')
        best_path = []
        for i, child in enumerate(node):
            val, path = minimax(child, depth-1, False, name + f"->C{i}")
            if val > best:
                best = val
                best_path = [i] + path
        print(f"{name} (Max) = {best}")
        return best, best_path

    else:
        best = float('inf')
        best_path = []
        for i, child in enumerate(node):
            val, path = minimax(child, depth-1, True, name + f"->C{i}")
            if val < best:
                best = val
                best_path = [i] + path
        print(f"{name} (Min) = {best}")
        return best, best_path



def alphabeta(node, depth, alpha, beta, isMax, name="Root"):
    if depth == 0:
        print(f"{name} = {node}")
        return node, []

    if isMax:
        best = float('-inf')
        best_path = []
        for i, child in enumerate(node):
            val, path = alphabeta(child, depth-1, alpha, beta, False, name + f"->C{i}")
            if val > best:
                best = val
                best_path = [i] + path
            alpha = max(alpha, best)

            if alpha >= beta:
                print(f"PRUNED at {name}")
                break

        print(f"{name} (Max) = {best}")
        return best, best_path

    else:
        best = float('inf')
        best_path = []
        for i, child in enumerate(node):
            val, path = alphabeta(child, depth-1, alpha, beta, True, name + f"->C{i}")
            if val < best:
                best = val
                best_path = [i] + path
            beta = min(beta, best)

            if alpha >= beta:
                print(f"PRUNED at {name}")
                break

        print(f"{name} (Min) = {best}")
        return best, best_path



print("\n--- MINIMAX ---")
value, path = minimax(tree, 3, True)

print("\nMinimax Value:", value)
print("Optimal Path (indices):", path)




print("\n--- ALPHA-BETA PRUNING ---")
value_ab, path_ab = alphabeta(tree, 3, float('-inf'), float('inf'), True)

print("\nAlpha-Beta Value:", value_ab)
print("Optimal Path (indices):", path_ab)
```

```python
# Task 4

%pip install ortools

from ortools.sat.python import cp_model

model = cp_model.CpModel()

A = model.NewIntVar(0, 3, 'A')
B = model.NewIntVar(0, 3, 'B')
C = model.NewIntVar(0, 3, 'C')

model.Add(A != B)
model.Add(B != C)
model.Add(A + B <= 4)

solver = cp_model.CpSolver()

status = solver.Solve(model)

print("Status:", status)  # DEBUG LINE

if status == cp_model.FEASIBLE or status == cp_model.OPTIMAL:
    print("A:", solver.Value(A))
    print("B:", solver.Value(B))
    print("C:", solver.Value(C))
else:
    print("No solution found")
```

```python
#task 5
class SolutionPrinter(cp_model.CpSolverSolutionCallback):
    def __init__(self, A, B, C):
        cp_model.CpSolverSolutionCallback.__init__(self)
        self.A = A
        self.B = B
        self.C = C

    def OnSolutionCallback(self):
        print(self.Value(self.A), self.Value(self.B), self.Value(self.C))

solver = cp_model.CpSolver()
solver.parameters.enumerate_all_solutions = True

solution_printer = SolutionPrinter(A, B, C)

solver.SearchForAllSolutions(model, solution_printer)
```

```python
#task6
model = cp_model.CpModel()

x = model.NewIntVar(0, 20, 'x')
y = model.NewIntVar(0, 20, 'y')
z = model.NewIntVar(0, 20, 'z')

model.Add(x + 2*y + z <= 20)
model.Add(3*x + y <= 18)

model.Maximize(4*x + 2*y + z)

solver = cp_model.CpSolver()
solver.Solve(model)

print("Optimal value:", solver.ObjectiveValue())
print("x:", solver.Value(x))
print("y:", solver.Value(y))
print("z:", solver.Value(z))
```

```python
model = cp_model.CpModel()

n = 4
queens = [model.NewIntVar(0, n-1, f'q{i}') for i in range(n)]

# No same column
model.AddAllDifferent(queens)

# No diagonal attacks
for i in range(n):
    for j in range(i+1, n):
        model.Add(queens[i] - queens[j] != i - j)
        model.Add(queens[i] - queens[j] != j - i)

solver = cp_model.CpSolver()
solver.Solve(model)

# Print board
for i in range(n):
    row = ['_'] * n
    row[solver.Value(queens[i])] = 'Q'
    print(" ".join(row))
```
